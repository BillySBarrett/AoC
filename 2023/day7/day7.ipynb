```python
# Reading the input file
with open('input.txt') as f:
    data = f.read().splitlines()

# Function to calculate counts of each card type
def get_counts(hand):
    cardType = '23456789TJQKA'
    counts = {r: 0 for r in cardType}
    for card in hand:
        counts[card[0]] += 1
    
    return counts

# Function to classify the hand
def classify(hand):
    counts = get_counts(hand)
    values = counts.values()

    if 5 in values:
        return 6
    elif 4 in values:
        return 5
    elif 3 in values and 2 in values:
        return 4
    elif 3 in values:
        return 3
    elif list(values).count(2) == 2:
        return 2
    elif 2 in values:
        return 1
    return 0  # high card

# Function to determine the strength of the hand
def strength(hand):
    return (classify(hand), hand)

# Parsing the input data and sorting
plays = []
for line in data:
    hand, value = line.split()
    plays.append((hand, int(value)))

plays.sort(key=lambda x: strength(x[0]))

# Calculating the total
total = 0
for rank, (hand, value) in enumerate(plays, 1):
    total += rank * value

print(plays)
print(total)
```

```python

letterMap = {'T':'A', 'Q':'B', 'K':'C', 'A':'D'}

def score(hand):
    counts = [hand.count(card) for card in hand]
    if 5 in counts:
        return 6
    elif 4 in counts:
        return 5
    elif 3 in counts and 2 in counts:
        return 4
    elif 3 in counts:
        return 3
    elif counts.count(2) == 4:
        return 2
    elif 2 in counts:
        return 1
    return 0

def replacements(hand):
    res = [hand]
    while True:
        for char in hand:
            if char == "J":
                for c in "23456789TQKA":
                    newHand = hand.replace(char, c)
                    res.append(newHand)
        return res

def classify(hand):
    return max(map(score, replacements(hand)))

def strength(hand):
    return (classify(hand), [letterMap.get(card, card) for card in hand])

plays = []
for line in open('input.txt'):
    hand, value = line.split()
    plays.append((hand, int(value)))
plays.sort(key=lambda x: strength(x[0]))

total = 0
for rank, (hand, value) in enumerate(plays, 1):
    total += (rank * value)

print(total)
```
